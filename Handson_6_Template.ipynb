## 12.1 Champion Counter
Create a program that reads a text file that contains a list of FIFA World Cup champions and determines the country that has won the most championships.

### Console:
```powershell
Country	        Wins	Years	
=======	        ====	=====	
Argentina       2       1978, 1986
Brazil	        5       1958, 1962, 1970, 1994, 2002
England	        1       1966		
France	        1       1998		
Germany	        4       1954, 1974, 1990, 2014
Italy	        4       1934, 1938, 1982, 2006
Spain	        1       2010
Uruguay	        2       1930, 1950
```

### Specifications:
- Use the attached text file named `world_cup_champions.txt` that contains data like this: 
```powershell
Year,Country,Coach,Captain 
1930,Uruguay,Alberto Suppici,JosÃ© Nasazzi 
1934,Italy,Vittorio Pozzo,Gianpiero Combi 
1938,Italy,Vittorio Pozzo,Giuseppe Meazza
...
```
- When the program starts, it should read the text file and use a dictionary to store the required data using the name of each country that has won the World Cup as the key.
- The program should compile the data shown above and display the countries alphabetically.



````python
"""
main
1. read file with function
open and read
first line goes out
split and append
2. display with function
key = country, value = year's'

3. choose max win
"""
def read_file(txt_file):
    dictionary = {}
    with open(txt_file, 'r') as file:
        firstline = file.readline()
        for line in file:
            line = line.strip()
            if not line:
                continue
            year, country, coach, captain = line.split(',')
            if country in dictionary:
                dictionary[country].append(year)
            else:
                dictionary[country] = [year]
    return dictionary

def display(dictionary):
    countries = sorted(dictionary.keys())
    print(f"Country\t        Wins\tYears")
    print(f"=======\t        ====\t=====")
    for country in countries:
        wins = len(dictionary[country])
        years = ",".join(dictionary[country])
        print(f"{country}\t        {wins}\t{years}")

def main():
    txt_file = "world_cup_champions.txt"
    dictionary = read_file(txt_file)
    display(dictionary)

    count_wins = 0
    count_country = ''
    for country, years in dictionary.items():
        if len(years) > count_wins:
            count_wins = len(years)
            count_country = country
    print()
    print(f"{count_country} has {count_wins} wins.")
    
if __name__ == '__main__':
    main()
````

## 12.2 Monthly Sales
Create a program that allows you to view and edit the sales amounts for each month of the current year.

### Console:

### Specifications:
- Use the attached text file named `monthly_sales.txt` that consists of rows that contain three-letter abbreviations for the month and the monthly sales.
- The program should read the file and store the sales data for each month in a dictionary with the month abbreviation as the key for each item.
- Whenever the sales data is edited, the program should write the changed data to the text file.




````python
### CODE HERE ###
def load_sales(filename):
    sales_data = {}
    with open(filename, "r") as file:
        for line in file:
            dic = line.split()
            if len(dic) >= 2:
                month = dic[0]
                sales_value = float(dic[1])
                sales_data[month] = sales_value
    return sales_data

def save_sales(filename, sales_data):
    with open(filename, "w") as file:
        for month, sales in sales_data.items():
            file.write(f"{month}\t{int(sales)}\n")

def view_sales(sales_data):
    month_input = input("Three-letter Month: ").strip()
    month_key = month_input.title() 
    if month_key in sales_data:
        print(f"Sales amount for {month_key} is {sales_data[month_key]:,.2f}.")
    else:
        print("Invalid three-letter month.")

def edit_sales(sales_data, filename):
    month_input = input("Three-letter Month: ").strip()
    month_key = month_input.title()
    if month_key in sales_data:
        amount_str = input("Sales Amount: ").strip()
        new_amount = float(amount_str)
        sales_data[month_key] = new_amount
        save_sales(filename, sales_data)
        print(f"Sales amount for {month_key} is {sales_data[month_key]:,.2f}.")
    else:
        print("Invalid three-letter month.")

def totals_sales(sales_data):
    total = sum(sales_data.values())
    average = total / len(sales_data) if sales_data else 0
    print(f"Yearly total: {total:,.2f}")
    print(f"Monthly average: {average:,.2f}")

def main():
    filename = "monthly_sales.txt"
    sales_data = load_sales(filename)

    print("Monthly Sales program\n")
    print("COMMAND MENU")
    print("view\t- View sales for specified month")
    print("edit\t- Edit sales for specified month")
    print("totals - View sales summary for year")
    print("exit\t- Exit program\n")

    while True:
        command = input("Command: ").strip()
        if command == "view":
            view_sales(sales_data)
        elif command == "edit":
            edit_sales(sales_data, filename)
        elif command == "totals":
            totals_sales(sales_data)
        elif command == "exit":
            print("Bye!")
            break
        else:
            print("Not a valid command. Please try again.\n")

if __name__ == "__main__":
    main()
````
